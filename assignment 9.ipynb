```python
import sympy as sp
```

```python
#code for a)
x, y = sp.symbols('x y')
# Define a function
f = sp.exp(x) * sp.sin(y) + y**3

# Differentiate f with respect to x
df_dx = sp.diff(f, x)

# Differentiate f with respect to y
df_dy = sp.diff(f, y)

print(df_dx)
print(df_dy)
```

```python
#code for b)
x, y = sp.symbols('x y')
# Define a function
g = x**2*y + x*y**2


grad = [sp.diff(g, var) for var in (x, y)]
print(grad)


# Evaluate at (1, -1)
grad_at_p = [expr.subs({x: 1, y: -1}) for expr in grad]
print(grad_at_p)

# Compute the magnitude
magnitude = sp.sqrt(sum(comp**2 for comp in grad_at_p))
print(magnitude)
print(magnitude.evalf())
```

```python
# code for d) (mostly straight from assignment)
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
